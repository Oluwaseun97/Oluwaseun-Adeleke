# TASK 1

```python

movie ={"Years":[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
        "durations": [103, 101, 99, 100,100, 95, 95, 96, 93, 90]}
movie
```

# TASK 2

```python
import pandas as pd
```

```python
df =pd.DataFrame(movie)
df
```

# TASK 3

```python
import matplotlib.pyplot as plt
```

```python
duration = df['durations']
duration
```

```python
year = df['Years']
year
```

```python
plt.figure(figsize=(10, 6))
plt.plot(year, duration, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Durations')
plt.title('Netflix Movies Durations 2011-2020')

plt.grid(True)  
plt.show()
```

# TASK 4

```python
netflix_df =pd.read_csv('netflix_data.csv')
netflix_df.head()
```

# TASK 5

```python
netflix_df_movies_only = netflix_df[netflix_df["type"]=="Movie"]
netflix_df_movies_only
```

```python
netflix_movies_col_subset = netflix_df_movies_only.loc[0:, ["title", 'country', "genre", "release_year", 'duration']]
netflix_movies_col_subset.head()
```

# TASK 6

```python
plt.figure(figsize=(10, 6))  
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color='b')

plt.xlabel('Release Year')
plt.ylabel('Movie Duration (minutes)')
plt.title('Movie Durations by Year of Release')

# for i, title in enumerate(movie_titles):
#     plt.text(years[i], durations[i], title, fontsize=8, ha='left', va='bottom', color='red')

plt.grid(True)  
plt.show()
```

# TASK 7

```python
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]
short_movies.head(20)
```

# TASK 8

```python
colours =[]
for colour in netflix_movies_col_subset["genre"]:
    if colour == "Children":
        colours.append("blue")
    elif colour == "Documentaries":
        colours.append("red")
    else:
        colours.append("black")
colours
```

# TASK 9

```python
plt.figure(figsize=(10, 6))  
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color=colours)

plt.xlabel('Release Year')
plt.ylabel('Movie Duration (minutes)')
plt.title('Movie Durations by Year of Release')


plt.grid(True)  
plt.show()
```
